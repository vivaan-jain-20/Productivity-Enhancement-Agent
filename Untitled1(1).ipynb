````python
# --- STEP 1: INSTALLATIONS & SETUP ---
# !pip install -U langgraph langchain-google-genai langchain-core

import os
import json
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict, List, Dict, Any
from langgraph.graph import StateGraph, START, END

# 1. Auth
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Paste your Google AI Studio Key: ")

# 2. Model (Gemini 1.5 Flash is great for this logic)
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0.0)

# 3. LONG-TERM MEMORY SETUP (The "Database")
MEMORY_FILE = "scheduler_memory.json"

def load_memory():
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, "r") as f:
            return json.load(f)
    return [] # Empty list if no memory yet

def save_memory(new_feedback):
    history = load_memory()
    history.append(new_feedback)
    with open(MEMORY_FILE, "w") as f:
        json.dump(history, f)
    print("💾 Agent has learned this preference for next time.")

# --- STEP 2: DEFINE STATE ---
class SchedulerState(TypedDict):
    tasks: List[Dict]
    calendar_slots: List[str]
    energy_map: Dict[str, str]
    schedule: Dict[str, Any]
    user_feedback: str  # New field to capture input

# --- STEP 3: DEFINE INTELLIGENT NODES ---

def infer_energy_node(state: SchedulerState):
    """
    Reads past user feedback to decide energy levels.
    """
    print("⚡ Agent is analyzing energy patterns & past memory...")
    slots = state["calendar_slots"]
    past_feedback = load_memory() # READ from "Database"

    # We use the LLM to apply your past constraints to the current slots
    prompt = f"""
    You are an Energy Analyzer.
    Map the following time slots to 'High Energy' or 'Low Energy'.

    DEFAULT LOGIC: Morning is High, Afternoon is Low.

    CRITICAL - USER PAST FEEDBACK (Override default if needed):
    {past_feedback}

    TIME SLOTS TO ANALYZE:
    {slots}

    Return a valid JSON object mapping slot -> energy level.
    Example: {{"9:00": "High Energy", "14:00": "Low Energy"}}
    """

    try:
        # Get JSON from LLM
        response = llm.invoke(prompt)
        # Clean the response to ensure it's valid JSON
        clean_json = response.content.replace("```json", "").replace("```", "").strip()
        energy_map = json.loads(clean_json)
    except:
        # Fallback if LLM fails
        energy_map = {s: "Medium Energy" for s in slots}

    return {"energy_map": energy_map}

def schedule_node(state: SchedulerState):
    """
    Assigns tasks based on the (potentially altered) energy map.
    """
    print("📅 Agent is matching tasks to slots...")
    tasks = sorted(state["tasks"], key=lambda x: x['priority'])
    energy_map = state["energy_map"]
    remaining_slots = state["calendar_slots"].copy()
    final_schedule = {}

    for task in tasks:
        if not remaining_slots: break

        # Priority 1 tasks want "High Energy" slots
        assigned_slot = remaining_slots[0]

        if task['priority'] == 1:
            for slot in remaining_slots:
                # We look for "High" in the description provided by the previous node
                if "High" in energy_map.get(slot, ""):
                    assigned_slot = slot
                    break

        final_schedule[assigned_slot] = task
        remaining_slots.remove(assigned_slot)

    return {"schedule": final_schedule}

def review_node(state: SchedulerState):
    """
    Displays schedule and CAPTURES user input.
    """
    print("\n--- 🤖 AGENT PROPOSED SCHEDULE ---")
    for time, task in state["schedule"].items():
        energy_context = state['energy_map'].get(time, "Unknown")
        print(f"[{time}] {task['name']} \t--> ({energy_context})")

    print("\n-----------------------------------")
    # THE INTERACTIVE PART
    feedback = input("Type your feedback (e.g., 'I hate 9am meetings' or 'Looks good'): ")
    return {"user_feedback": feedback}

def learning_node(state: SchedulerState):
    """
    Saves the feedback to long-term memory.
    """
    feedback = state["user_feedback"]

    # If the user provided meaningful feedback (not just 'ok'), save it
    if len(feedback) > 3 and "good" not in feedback.lower() and "ok" not in feedback.lower():
        save_memory(feedback)
    else:
        print("👍 No new constraints to learn.")

    return {}

# --- STEP 4: BUILD GRAPH ---
workflow = StateGraph(SchedulerState)

workflow.add_node("energy_analyzer", infer_energy_node)
workflow.add_node("scheduler", schedule_node)
workflow.add_node("human_review", review_node)
workflow.add_node("learner", learning_node) # New Node

workflow.add_edge(START, "energy_analyzer")
workflow.add_edge("energy_analyzer", "scheduler")
workflow.add_edge("scheduler", "human_review")
workflow.add_edge("human_review", "learner")
workflow.add_edge("learner", END)

app = workflow.compile()

# --- STEP 5: RUN THE AGENT (Run this cell twice to test!) ---

print("🚀 Starting Scheduler Agent...")

# Data
input_data = {
    "tasks": [
        {"name": "Deep Coding", "priority": 1, "duration": 2},
        {"name": "Email Catchup", "priority": 2, "duration": 1},
    ],
    "calendar_slots": ["9:00", "10:00", "13:00", "14:00"],
    "energy_map": {}, "schedule": {}, "user_feedback": ""
}

result = app.invoke(input_data)
````

````python
# --- STEP 1: INSTALLATIONS ---
# Run this once in your terminal or Colab cell
# !pip install -U langchain-google-genai langgraph langchain-core

import os
import getpass
from typing import TypedDict, Dict, Any
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.graph import StateGraph, START, END

# --- STEP 2: SETUP API & MODEL ---
# 1. Securely set the API Key
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Paste your Google AI Studio Key: ")

# 2. Initialize the LLM (Using 2.5 Flash for speed)
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash", # Change to "gemini-2.0-flash-exp" if you have beta access
    temperature=0.0
)

# --- STEP 3: DEFINE STATE ---
class EmailButlerState(TypedDict):
    email_content: Dict[str, str]
    is_spam: bool
    detected_style: str
    draft_response: str
    execution_log: str

# --- STEP 4: DEFINE INTELLIGENT NODES ---

def triage_officer_node(state: EmailButlerState):
    """
    Node 1: Uses LLM to intelligently detect spam (better than keyword lists).
    """
    print("🔍 Triage Officer: Asking Gemini to analyze intent...")
    email = state["email_content"]

    # AI Prompt for Classification
    prompt = f"""
    Analyze this email. Is it SPAM or promotional junk?
    Return ONLY the word 'TRUE' or 'FALSE'.

    Subject: {email.get('subject')}
    Body: {email.get('body')}
    """
    response = llm.invoke(prompt)

    # Parse the AI's answer
    is_spam_bool = "TRUE" in response.content.upper()
    return {"is_spam": is_spam_bool}

def style_mimic_node(state: EmailButlerState):
    """
    Node 2: Uses LLM to analyze the sender's tone.
    """
    print("✍️ Style Mimic: Analyzing communication style...")
    body = state["email_content"].get("body")

    prompt = f"""
    Read this email and describe the writer's tone in 5 words or less.
    (e.g., 'Formal and urgent', 'Casual and technical').

    Email Body: {body}
    """
    response = llm.invoke(prompt)
    return {"detected_style": response.content.strip()}

def draft_architect_node(state: EmailButlerState):
    """
    Node 3: Generates a contextual reply.
    """
    print("📝 Draft Architect: Generating response...")
    email = state["email_content"]
    style = state["detected_style"]

    prompt = f"""
    You are a helpful assistant. Write a reply to this email.

    CONTEXT:
    - Sender: {email.get('sender')}
    - Topic: {email.get('subject')}
    - Tone to use: {style}

    Reply:
    """
    response = llm.invoke(prompt)
    return {"draft_response": response.content}

def sql_adaptation_node(state: EmailButlerState):
    """
    Node 4: Logging (unchanged logic, just reporting AI results).
    """
    print("💾 SQL Agent: Logging transaction...")
    status = "BLOCKED (Spam)" if state["is_spam"] else "PROCESSED"
    return {"execution_log": f"Status: {status}"}

# --- STEP 5: BUILD THE GRAPH ---
workflow = StateGraph(EmailButlerState)

# Add Nodes
workflow.add_node("triage", triage_officer_node)
workflow.add_node("mimic", style_mimic_node)
workflow.add_node("architect", draft_architect_node)
workflow.add_node("logger", sql_adaptation_node)

# Add Conditional Edges
def routing_logic(state: EmailButlerState):
    if state["is_spam"]:
        print("   --> Routing to Logger (Spam Detected)")
        return "logger"
    else:
        print("   --> Routing to Mimic (Legitimate Email)")
        return "mimic"

workflow.add_edge(START, "triage")
workflow.add_conditional_edges("triage", routing_logic)
workflow.add_edge("mimic", "architect")
workflow.add_edge("architect", "logger")
workflow.add_edge("logger", END)

app = workflow.compile()

# --- STEP 6: RUN DEMO ---

def run_demo(email_dict, description):
    print(f"\n--- DEMO: {description} ---")
    initial_state = {
        "email_content": email_dict,
        "is_spam": False, "detected_style": "", "draft_response": "", "execution_log": ""
    }

    result = app.invoke(initial_state)

    print(f"\n📊 FINAL LOG: {result['execution_log']}")
    if result['draft_response']:
        print(f"📧 GENERATED DRAFT:\n{'-'*20}\n{result['draft_response']}\n{'-'*20}")

# Example 1: Technical Query
legit_email = {
    "sender": "John (Lab Partner)",
    "subject": "Matrix Pointers help",
    "body": "Hey, I'm stuck on the pointer arithmetic for the matrix multiplication lab. Can you peek at my code?"
}

# Example 2: Obvious Spam
spam_email = {
    "sender": "Prize Bot",
    "subject": "YOU WON $1,000,000",
    "body": "Click here immediately to claim your massive cash prize!!!"
}

# Run
run_demo(legit_email, "Legitimate Request")
run_demo(spam_email, "Spam Attack")
````

````python
# --- STEP 1: INSTALLATIONS ---
# Run this once. -U ensures you have the latest versions compatible with each other.
!pip install -U langgraph langchain-google-genai langchain-core

# --- STEP 2: SETUP & AUTHENTICATION (Fixes ValidationError) ---
import os
import getpass
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict, List, Dict, Any
from langgraph.graph import StateGraph, START, END

# 1. Securely enter your Google API Key
if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Paste your Google AI Studio Key: ")

# 2. Initialize the Model (Now it will work because the key is set!)
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)

# --- STEP 3: DEFINE AGENT STATE ---
class TravelState(TypedDict):
    request: str            # User input
    itinerary: str          # Day-by-day plan
    budget: str             # Cost estimate
    final_doc: str          # Final output

# --- STEP 4: DEFINE NODES ---

def planner_node(state: TravelState):
    print("✈️  Designing Itinerary...")
    prompt = f"Create a 3-day itinerary for: {state['request']}. Be specific."
    response = llm.invoke(prompt)
    return {"itinerary": response.content}

def budget_node(state: TravelState):
    print("💰 Estimating Costs...")
    prompt = f"Create a rough budget (Flights, Hotel, Food) for this trip:\n\n{state['itinerary']}"
    response = llm.invoke(prompt)
    return {"budget": response.content}

def printer_node(state: TravelState):
    print("🖨️  Compiling Travel Pack...")
    doc = f"--- 🌍 TRIP PLAN ---\n\n{state['itinerary']}\n\n--- 💵 BUDGET ---\n\n{state['budget']}"
    return {"final_doc": doc}

# --- STEP 5: COMPILE & RUN ---
workflow = StateGraph(TravelState)
workflow.add_node("planner", planner_node)
workflow.add_node("budget", budget_node)
workflow.add_node("printer", printer_node)

workflow.add_edge(START, "planner")
workflow.add_edge("planner", "budget")
workflow.add_edge("budget", "printer")
workflow.add_edge("printer", END)

app = workflow.compile()

# Run it
print("\n🚀 Launching Travel Agent...")
inputs = {"request": "5 days in Kyoto, Japan. History and Sushi.", "itinerary": "", "budget": "", "final_doc": ""}
result = app.invoke(inputs)

print(result["final_doc"])
````
